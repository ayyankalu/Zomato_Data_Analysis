**Zomato Data Analysis**

```python
#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating the dataframe
dataframe = pd.read_csv("/content/Zomato data .csv")
print(dataframe)
```

```python
#Convert the datatype of Rate
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)
dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.head()
```

**Type of Restaurant**

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of restaurant')
plt.ylabel('Count')
```

Conclusion - Majority of the restaurant falls under dining category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel('Type of restaurant', c="red", size=20)
plt.ylabel('Votes', c="blue", size=20)
```

Conclusion - dining restaurants has received maximum votes


```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'], bins = 5)
plt.title("Ratings distribution")
plt.show()
```

Conclusion - The majority restaurants received ratings between 3.5 to 4

```python
dataframe.head()
```

**Average order spending by couples**

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
plt.xlabel('Cost')
plt.ylabel('Count')
```

Conclusion - the mojrity of couples prefer restaurants with an approx cost of Rs.300

**Which mode receives the maximum rating**

```python
dataframe.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

Conclusion - Offline order recived lower rating as compared to online order


```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True, fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in Type")
plt.show()
```

Conclusion - Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at restaurants, but prefer online ordering in cafes.
